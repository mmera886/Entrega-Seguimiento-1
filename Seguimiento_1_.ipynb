```python
 #Ejercicio numero 4#
edadjuan = 9
edadalberto = (2/3) * edadjuan
edadana = (4/3) * edadjuan
edadmama = edadjuan + edadalberto + edadana
print("La edad de juan es:", edadjuan)
print("La edad de ana es:", edadana)
print("La edad de alberto es:", edadalberto)
print("La edad de ana es de:", edadana)
print("La edad de la mama es de:", edadmama)
```

```python
#Ejercicio numero 12#
horas = 48
precio_hora = 5000
porcentaje_retencion = 0.125
salario_bruto = precio_hora * horas
valor_retencion = salario_bruto * porcentaje_retencion
salario_neto = salario_bruto - valor_retencion
print("El valor del salario bruto es de:", salario_bruto)
print("El valor de la retencion es de:", valor_retencion)
print("El valor del salario neto es de:", salario_neto)



```
